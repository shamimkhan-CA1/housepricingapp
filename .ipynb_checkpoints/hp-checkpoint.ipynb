## Data pre-processing

```python
#label encode - encode categorical values into number by assigning 0 to and onwards
```

```python
#!pip install scikit-learn #largest library in python for Text base Machine Learning activities
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder() #creating object
```

```python
le1=LabelEncoder() #creating object
```

```python
df['Location']=le.fit_transform(df['Location'])
```

```python
df['Condition']=le1.fit_transform(df['Condition'])
```

```python
df
```

```python
le.classes_
```

```python
le1.classes_
```

```python
le2=LabelEncoder() #creating object
```

```python
df['Garage']=le2.fit_transform(df['Garage'])
```
